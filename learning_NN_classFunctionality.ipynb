```python
import torch
import torch.nn as nn
```

```python
#sequential flow
#only defining layer in constructor
class model(nn.Module):
  def __init__(self,features):
    super().__init__()

    self.network=nn.Sequential(
        nn.Linear(features,3),
        nn.ReLU(),
        nn.Linear(3,1),
        nn.Sigmoid()
    )
  def forward(self,features):
    return self.network(features)

# create dataset
features = torch.rand(10,5)

# create model
model = model(features.shape[1])

# call model for forward pass
# model.forward(features)#<-1st
#model(features)#<-2nd
#to view weights
model.network[0].weight    # First Linear layer
```

```python
#non sequential flow
#defining eacH LAYER in constructor and forward function both
class model1(nn.Module):
  def __init__(self,features):
    super().__init__()

    self.linear1=nn.Linear(features,3)
    self.relu=nn.ReLU()
    self.linear2=nn.Linear(3,1)
    self.sigmoid=nn.Sigmoid()

  def forward(self,features):
    x=self.linear1(features)
    x=self.relu(x)
    x=self.linear2(x)
    x=self.sigmoid(x)
    return x

# create dataset
features = torch.rand(10,5)

# create model
model1 = model1(features.shape[1])

# call model for forward pass
# model.forward(features)#<-1st
#model(features)#<-2nd
#to view weights
model1.linear1.weight    # First Linear layer
```
